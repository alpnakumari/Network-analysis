### Most Influential
+ nx.degree_centrality()

```python
nx.degree_centrality(criminal_graph)
```

```python
sorted(nx.degree_centrality(criminal_graph).values())
```

```python
most_influential = nx.degree_centrality(criminal_graph)
```

```python
for w in sorted(most_influential, key=most_influential.get, reverse=True):
    print(w, most_influential[w])
```

# CLOSENESS CENTRALITY
 + This is based on the assumption that important nodes are close to other nodes. It is calculated as the sum of the path      lengths from the given node to all other nodes.

```python
sorted(nx.closeness_centrality(criminal_graph).values())
```

```python
important_node = nx.closeness_centrality(criminal_graph)
for w in sorted(important_node, key=important_node.get, reverse=True):
    print(w, important_node[w])
```

### Most Important Connection
 + nx.eigenvector_centrality()

```python
nx.eigenvector_centrality(criminal_graph)
```

```python
most_important_link = nx.eigenvector_centrality(criminal_graph)
```

```python
for w in sorted(most_important_link, key=most_important_link.get, reverse=True):
    print(w, most_important_link[w])
```

```python
nx.shortest_path(criminal_graph,"Hani Hanjour","Hamza Alghamdi")
```

```python
nx.shortest_path(criminal_graph,"Mohamed Atta","Satam Suqami")
```

```python
nx.shortest_path(criminal_graph,"Mohamed Abdi","Osama Awadallah")
```

### Betweenness Centrality
  + Bridge or Connection (Betweeness centrality unifies the network).
  + It's a measure of how often a node is a bridge between the nodes.

```python
plt.figure(figsize=(25,20))
nx.draw_networkx(criminal_graph)
plt.show()
```

```python
nx.betweenness_centrality(criminal_graph)
```

```python
best_mediator = nx.betweenness_centrality(criminal_graph)
for w in sorted(best_mediator, key=best_mediator.get, reverse=True):
    print(w, best_mediator[w])
```
